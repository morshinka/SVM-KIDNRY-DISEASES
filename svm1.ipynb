```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import  svm
from sklearn.metrics import  accuracy_score
```

```python
df = pd.read_csv('fixed_data.csv')
```

```python
df.sample()
```

```python
x = df.drop(columns='classification',axis=1)
y = df['classification']
```

```python
print(x)
```

```python
print(y)
```

```python
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, stratify=y, random_state=2)
```

```python
clasifier = svm.SVC(kernel='linear')
```

```python
clasifier.fit(x_train, y_train)
```

```python
x_train_prediction = clasifier.predict(x_train)
train_accuracy = accuracy_score(x_train_prediction, y_train)
print('akurasi data latih : ', train_accuracy * 100)
```

```python
x_test_prediction = clasifier.predict(x_test)
test_accuracy = accuracy_score(x_test_prediction, y_test)
print('akurasi data testing :', test_accuracy * 100)
```
